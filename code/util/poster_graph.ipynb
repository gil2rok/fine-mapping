```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
cavi_file = open('/home/gil2rok/Downloads/ld1_cavi.csv', 'rb') # rewrite this file path
adam_file = open('/home/gil2rok/Downloads/ld1_opt.csv', 'rb') # rewrite this file path too

cavi = np.loadtxt(cavi_file, delimiter=',')
adam = np.loadtxt(adam_file, delimiter=',')
```

```python
plt.rcParams["figure.figsize"] = (12,7) # change figure size here with (width, height) parameters
plt.plot(adam[:,0], adam[:,1], label='Adam')
plt.plot(cavi[:,0], cavi[:,1], label='CAVI')
plt.legend()
plt.yscale('log')
plt.xlabel('Iteration')
plt.ylabel('ELBO')
plt.title('ELBO for CAVI and Adam')
plt.show()
```
